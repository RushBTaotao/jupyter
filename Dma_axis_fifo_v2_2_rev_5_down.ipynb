```python
from pynq import Overlay
overlay = Overlay("/home/xilinx/pynq/overlays/test/dma/design_1.bit")
import pynq.lib.dma
dma=overlay.axi_dma_0
dma_hls=overlay.axi_dma_1
from pynq import Xlnk
import numpy as np

data_size = 200

xlnk = Xlnk()

input_buffer = xlnk.cma_array(shape=(data_size,), dtype=np.uint32)
for i in range(data_size):
    input_buffer[i] = i + 0xcaf70000
print("Print first few values of input buffer ...")
for i in range(data_size):
    print(hex(input_buffer[i]))
dma.sendchannel.transfer(input_buffer)

output_buffer = xlnk.cma_array(shape=(data_size,), dtype=np.uint32)
dma.recvchannel.transfer(output_buffer)
print("Print first few values output")
for i in range(data_size):
    print('0x' + format(output_buffer[i], '02x'))
```

```python
from pynq import Overlay
overlay = Overlay("/home/xilinx/pynq/overlays/test/dma/design_1.bit")
import pynq.lib.dma
dma=overlay.axi_dma_0
dma_hls=overlay.axi_dma_1
from pynq import Xlnk
import numpy as np

data_size = 200

xlnk = Xlnk()

input_buffer = xlnk.cma_array(shape=(data_size,), dtype=np.uint32)
for i in range(data_size):
    input_buffer[i] = i + 0xcaf70000
print("Print first few values of input buffer ...")
for i in range(data_size):
    print(hex(input_buffer[i]))
dma_hls.sendchannel.transfer(input_buffer)

output_buffer = xlnk.cma_array(shape=(data_size,), dtype=np.uint32)
dma_hls.recvchannel.transfer(output_buffer)
print("Print first few values output")
for i in range(data_size):
    print('0x' + format(output_buffer[i], '02x'))
```

```python
for i in range(data_size):
    input_buffer[i] = input_buffer[i]+0x10000

output_buffer = xlnk.cma_array(shape=(data_size,), dtype=np.uint32)
dma.sendchannel.transfer(input_buffer)
dma.recvchannel.transfer(output_buffer)
dma.sendchannel.wait()
dma.recvchannel.wait()
for i in range(data_size):
    print('0x' + format(output_buffer[i], '02x'))
errorcount=0
for i in range(data_size):
   if input_buffer[i]!=output_buffer[i]:
    errorcount+=1

print("tranmission error rate %2.3f: " %(errorcount/data_size))
```
